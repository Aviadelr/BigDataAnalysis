The global clustering coefficient is defined [here](https://en.wikipedia.org/wiki/Clustering_coefficient), as 3 times the number of triangles, divided by the number of triplets, where a triplet is defined as a set of two edges ${(x,y),(x,z)}$ centered at a joint vertex $x$. Alternatively, a formula for the global clustering coefficient in terms of the adjacency matrix $E$ and the degrees vector $d$ is:
$$
\frac{\sum_{i,j,k=1}^n e_{ij} e_{ik} e_{jk}}{\sum_{i=1}^n d_i(d_i-1)}
$$

```python
in_degrees = G_wiki.in_degree()
out_degrees = G_wiki.out_degree()

top_5_out_degrees = sorted(out_degrees, key=lambda x: x[1], reverse=True)[:5]
print("Top 5 nodes by out-degree:")
for node, degree in top_5_out_degrees:
    print(f"Node ID {node}: Degree values {degree}")

top_5_in_degrees = sorted(in_degrees, key=lambda x: x[1], reverse=True)[:5]
print("\nTop 5 nodes by in-degree:")
for node, degree in top_5_in_degrees:
    print(f"Node ID {node}: Degree values {degree}")

in_degrees_values = list(dict(in_degrees).values())
out_degrees_values = list(dict(out_degrees).values())

in_degree_distribution = pd.Series(in_degrees_values).value_counts() / num_nodes
out_degree_distribution = pd.Series(out_degrees_values).value_counts() / num_nodes

bottom_5 = in_degree_distribution.quantile(0.05)
top_5 = in_degree_distribution.quantile(0.95)
filtered_degrees = in_degree_distribution[(in_degree_distribution >= bottom_5) & (in_degree_distribution <= top_5)]

log_degrees_in = np.log(filtered_degrees.index)
log_frequencies_in = np.log(filtered_degrees.values)

alpha_in, intercept_in, r_value, p_value, std_err = linregress(log_degrees_in, log_frequencies_in)

bottom_5 = out_degree_distribution.quantile(0.05)
top_5 = out_degree_distribution.quantile(0.95)
filtered_degrees = out_degree_distribution[(out_degree_distribution >= bottom_5) & (out_degree_distribution <= top_5)]

log_degrees_out = np.log(filtered_degrees.index)
log_frequencies_out = np.log(filtered_degrees.values)

alpha_out, intercept_out, r_value, p_value, std_err = linregress(log_degrees_out, log_frequencies_out)

print(" ")
print(f"The alpha for the in-degrees: {intercept_in}")
print(" ")
print(f"The alpha for the out-degrees: {intercept_out}")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Scatter plot for in-degrees
ax1.scatter(log_degrees_in, log_frequencies_in, label="Data")
ax1.plot(log_degrees_in, alpha_in * log_degrees_in + intercept_in, 'r', label="Regression Line")
ax1.set_xlabel("log(Degrees)")
ax1.set_ylabel("log(Frequency)")
ax1.set_title("In-Degrees Scatter Plot")
ax1.legend()

# Scatter plot for out-degrees
ax2.scatter(log_degrees_out, log_frequencies_out, label="Data")
ax2.plot(log_degrees_in, alpha_out * log_degrees_in + intercept_out, 'r', label="Regression Line")
ax2.set_xlabel("log(Degrees)")
ax2.set_ylabel("log(Frequency)")
ax2.set_title("Out-Degrees Scatter Plot")
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
# Q8. Code
import random
import copy

def sample_random_graph_with_degree_seq(G_init, iters = 100000):
    G_rand = G_init # Note! we don't do deep copy here to save time. But the function can modify G_init. It is better to keep a separate copy of G outside the function
    edges = random.sample(list(G_rand.edges()), 2*iters)

    for i in range(0, iters, 2):
#        if i%1000 == 0:
#            print("Run: ", i)  # you can turn off/on internal printing
        if edges[i] in G_rand.edges() and edges[i+1] in G_rand.edges() and len(set(edges[0] + edges[1])) == 4: # swap
            G_rand.remove_edges_from([edges[i], edges[i+1]])
            G_rand.add_edges_from([ (edges[i][0], edges[i+1][1]), (edges[i][1], edges[i+1][0])  ])

    return G_rand

```

```python
random_networks = []

# set the initial input graph to G_wiki_undirected
G = G_wiki_undirected.copy()
for i in range(100):

    # generate a random network using sample_random_graph_with_degree_seq
    G_rand = sample_random_graph_with_degree_seq(G,50000)

    # add the generated random network to the list
    random_networks.append(G_rand)

    # set the input graph for the next iteration to the generated random network
    G = copy.deepcopy(G_rand)

clustering_coeffs = [nx.transitivity(G) for G in random_networks]

plt.figure(figsize=(12, 8))
plt.hist(clustering_coeffs, bins=10, edgecolor='black')
plt.xlabel('Clustering Coefficient')
plt.ylabel('Count')
plt.show()
```
